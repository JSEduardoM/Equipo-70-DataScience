# Limpieza

### Renombrar variables

```python
datos.rename(columns={'Age': 'Edad', 'Gender': 'Genero', 'Annual_Income':'Ingreso_Anual',
                      'Total_Spend':'Total_Gastado', 'Years_as_Customer':'Anios_como_cliente',
                      'Num_of_Purchases':'Numeros_Compras', 
                      'Average_Transaction_Amount':'Importe_Promedio_Transaccion',
                      'Num_of_Returns': 'Numero_de_Devoluciones',
                      'Num_of_Support_Contacts':'Cantidad_de_Contactos_Soporte',
                      'Satisfaction_Score': 'Nivel_Satisfaccion',
                      'Last_Purchase_Days_Ago':'Ultimo_dia_compra',
                      'Email_Opt_In':'Esta_Suscripto_via_Email',
                      'Promotion_Response':'Respuesta_a_Promocion',
                      'Target_Churn':'Target'}, inplace=True)
```

```python
datos.head()
```

### Cambio de tipo de variable

```python
datos['Genero'].astype('string')
```

```python
datos['Respuesta_a_Promocion'].astype('string')
```

# Analisis Univariado

```python

variables_numericas = [
    'Edad', 'Ingreso_Anual', 'Total_Gastado', 'Anios_como_cliente',
    'Numeros_Compras', 'Importe_Promedio_Transaccion', 'Numero_de_Devoluciones',
    'Cantidad_de_Contactos_Soporte', 'Nivel_Satisfaccion', 'Ultimo_dia_compra'
]
```

```python
plt.figure(figsize=(14, 10))
datos[variables_numericas].hist(bins=20, layout=(4, 3), figsize=(14, 10))
plt.tight_layout()
plt.show()
```

# Correlacin

```python
# Variables de comportamiento del cliente


variables_comportamiento = [
    'Numeros_Compras',
    'Importe_Promedio_Transaccion',
    'Total_Gastado',
    'Numero_de_Devoluciones',
    'Cantidad_de_Contactos_Soporte',
    'Nivel_Satisfaccion',
    'Ultimo_dia_compra',
    'Esta_Suscripto_via_Email',
    'Target'
]
```

```python
corr = datos[variables_comportamiento].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlaciones  Variables de Comportamiento")
plt.show()
```

# Anlisis adicional de relaciones

```python
# Relacin compras vs gasto total segn churn
sns.scatterplot(
    data=datos,
    x='Numeros_Compras',
    y='Total_Gastado',
    hue='Target'
)
plt.title("Relacin entre Nmero de Compras y Total Gastado (segn churn)")
plt.show()
```

# Nivel de satisfaccin vs churn

```python
sns.boxplot(data=datos, x='Target', y='Nivel_Satisfaccion')
plt.title("Nivel de Satisfaccin segn Target (Churn)")
plt.show()
```

# Respuesta a promociones vs churn

```python
sns.countplot(data=datos, x='Respuesta_a_Promocion', hue='Target')
plt.title("Respuesta a Promociones y su relacin con Churn")
plt.show()
```
