### Renombrar variables

```python
datos.rename(columns={
    'order_id': 'ID_Orden',
    'customer_id': 'ID_Cliente',
    'age': 'Edad',
    'product_id': 'ID_Producto',
    'country': 'Pais',
    'signup_date': 'Fecha_Registro',
    'last_purchase_date': 'Fecha_Ultima_Compra',
    'cancellations_count': 'Cantidad_Cancelaciones',
    'subscription_status': 'Estado_Suscripcion',
    'order_date': 'Fecha_Orden',
    'unit_price': 'Precio_Unitario',
    'quantity': 'Cantidad',
    'purchase_frequency': 'Frecuencia_Compra',
    'preferred_category': 'Categoria_Preferida',
    'product_name': 'Nombre_Producto',
    'category': 'Categoria',
    'gender': 'Genero'
}, inplace=True)
```

### Conversi贸n de fechas

```python
datos['Fecha_Registro'] = pd.to_datetime(datos['Fecha_Registro'])
datos['Fecha_Ultima_Compra'] = pd.to_datetime(datos['Fecha_Ultima_Compra'])
datos['Fecha_Orden'] = pd.to_datetime(datos['Fecha_Orden'])
```

### Datos Duplicados

```python
datos.duplicated().sum()
```

### Crear variable de Total de Orden

```python
datos['Total_Orden'] = datos['Precio_Unitario'] * datos['Cantidad']
```

### Crear variable Target basada en Estado_Suscripcion

# cancelled = 1 (churn), active/paused = 0 (no churn)

```python
datos['Target'] = (datos['Estado_Suscripcion'] == 'cancelled').astype(int)
```

# Analisis Univariado

### Selecci贸n de variables relevantes de comportamiento

```python
variables_comportamiento = [
    "Edad",
    "Cantidad_Cancelaciones",
    "Frecuencia_Compra",
    "Precio_Unitario",
    "Cantidad",
    "Total_Orden",
    "Target"
]
```

# Correlaci贸n

## Matriz de correlaci贸n

```python
corr = datos[variables_comportamiento].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlaciones  Variables de Comportamiento")
plt.show()

corr
```

## Separar variables num茅ricas y categ贸ricas del subset de comportamiento

```python
datos_comportamiento = datos[variables_comportamiento]

numericas = datos_comportamiento.select_dtypes(include=['int64', 'float64'])
categoricas = datos_comportamiento.select_dtypes(include=['object', 'string'])

print("Variables num茅ricas de comportamiento:", numericas.columns.tolist())
print("Variables categ贸ricas de comportamiento:", categoricas.columns.tolist())
```

## Histogramas para variables num茅ricas

```python
for col in numericas.columns:
    plt.figure(figsize=(7,4))
    sns.histplot(datos_comportamiento[col], kde=True)
    plt.title(f"Distribuci贸n de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```

## Boxplots para identificar outliers

```python
for col in numericas.columns:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=datos_comportamiento[col])
    plt.title(f"Boxplot de {col}")
    plt.show()
```
